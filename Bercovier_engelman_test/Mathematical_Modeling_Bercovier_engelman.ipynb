<div style="background-color: #e0f7fa; border-radius: 10px; width:90%;">
    <h2 style="color: #00796b;">Bercovier-Engelman Test</h2>
</div>

We consider the steady Stokes problem defined by the following equations:
$$
\begin{align}
    - \mu \Delta \mathbf{u} + \nabla p &= \mathbf{f} \quad \text{on} \quad \Omega, \\
    \nabla \cdot \mathbf{u} &= 0 \quad \text{on} \quad \Omega, \\
    \mathbf{u} &= 0 \quad \text{in} \quad \partial \Omega, \\
    %\frac{\partial \mathbf{u}}{\partial n} - p \cdot \mathbf{n} &= \mathbf{g} \quad \text{sur} \quad \Gamma_{\text{in}} \cup \Gamma_{\text{out}} = \{0,1\} \times [0,1].
\end{align}
$$

The exact solution of Bercovier-Engelman is defined on $\Omega=[ 0, 1 ]^2$ by:

$$
\begin{aligned}
u_1 &= -256y(y - 1)(2y - 1)x^2(x - 1), \\
u_2 &= 256x(x - 1)(2x - 1)y^2(y - 1), \\
p &= (x - 0.5)(y - 0.5)
\end{aligned}
$$

$%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%$

<div style="background-color: #e0f7fa; border-radius: 10px; width:90%;">
    <h2 style="color: #00796b;">Variational Forme</h2>
</div>

We set $ V= [H^{1}_{0}(\Omega)]^2 $ and $ Q=L^{2}_{0}(\Omega)$ =$\{\mathbf{v} \in L^2(\Omega), \quad \int_{\Omega} \mathbf{v}=0 \} $ We then multiply the equation (1) by $\mathbf{v} \in V$ and the equation (2) by $\mathbf{q}\in Q$, then integrate the equations over $\Omega$ and by Applying the Greenâ€™s formula, we get: 

Find $(\mathbf{u}, \mathbf{p}) \in \mathcal{W} = \mathcal{V} \times \mathcal{Q}$ such that:

$$
\begin{aligned}
     \mu \int_{\Omega} \nabla \mathbf{u} \cdot \nabla \mathbf{v} \, dx - \int_{\Omega} \mathbf{p} \nabla \cdot \mathbf{v} \, dx &= \int_{\Omega} \mathbf{f} \cdot \mathbf{v} \, dx &&, \quad \forall \mathbf{v} \in V, \\
     \int_{\Omega} \mathbf{q} \nabla \cdot \mathbf{u} \, dx &= 0 &&, \quad \forall \mathbf{q} \in Q.
\end{aligned}
$$

$%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%$

<div style="background-color: #e0f7fa; border-radius: 10px; width:90%;">
    <h2 style="color: #00796b;">Taylor-Hood spaces</h2>
</div>

Let $V_{h} \subset V$ and $Q_{h} \subset Q$ be subspaces of inite dimensions such as $V_{h} \to V$ and $Q_{h} \to Q$ when $h \to 0$. we define the Taylor-Hood space in the following way:

$$V_{h}=\{X^{2}_{h} \cup H^{1}_{0}(\Omega)\}^2$$
$$Q_{h}=\{X^{1}_{h} \cup L^{2}_{0}(\Omega) \}$$
Where $X^{r}_{h}=\{ g \in C(\bar{\Omega}), \quad K \in T_{h}, \quad  g|_{K} \in \mathbb{P}^{r} \}$

We consider the approximated Stokes problem: 

$$
\begin{aligned}
    \mu \int_{\Omega} \nabla \mathbf{u_{h}} \cdot \nabla \mathbf{v_{h}} \, dx - \int_{\Omega} \mathbf{p_{h}} \nabla \cdot \mathbf{v_{h}} \, dx &= \int_{\Omega} \mathbf{f} \cdot \mathbf{v_{h}} \, dx &&, \quad \forall \mathbf{v_{h}} \in V_{h}, \\
    \int_{\Omega} \mathbf{q_{h}} \nabla \cdot \mathbf{u_{h}} \, dx &= 0 &&, \quad \forall \mathbf{q_{h}} \in Q.
\end{aligned}
$$

This problem is well-defined if the spaces Vh and Qh verify the inf-sup condition:

$$
\exists \beta_h > 0, \quad \inf_{q_h \in Q_h} \sup_{v_h \in V_h} \frac{\int_{\Omega} q_h \, \text{div}(v_h)}{\|q_h\|_Q \|v_h\|_V} \geq \beta_h
$$


The problem is equivalent to the matricial form:

$$\begin{bmatrix}
\mathbb{A}_{h} & \mathbb{B}_{h}^t\\
\mathbb{B}_{h} & 0
\end{bmatrix}
\begin{bmatrix}
u_h\\
p_h
\end{bmatrix}
=
\begin{bmatrix}
\mathbb{F}_h\\
0
\end{bmatrix}$$

Where $\mathbb{A}_{H}$, $\mathbb{B}_{H}$ and $\mathbb{F}_{H}$ are the respectives representatives matrices of the bilinear forms:

$$ a(\mathbf{u_{h}},\mathbf{v_{h}})=\mu \int_{\Omega} \nabla \mathbf{u_{h}} \cdot \nabla \mathbf{v_{h}} \, dx $$ 
$$ b(\mathbf{p_{h}},\mathbf{v_{h}})=\int_{\Omega} \mathbf{p_{h}} \nabla \cdot \mathbf{v_{h}} \, dx $$
$$ f(\mathbf{v_{h}})=\int_{\Omega} \mathbf{f} \cdot \mathbf{v_{h}} \, dx $$

$%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%$

<div style="background-color: #e0f7fa; border-radius: 10px; width:90%;">
    <h2 style="color: #00796b;">convergence estimation</h2>
</div> 

The Taylor-Hood spaces verify the inf-sup conditions, we have the convergence estimation:

$$
\lVert u - u_h \rVert_{\mathcal{V}} + \lVert p - p_h \rVert_{\mathcal{Q}} \leq C h^{\ell} \left( \lVert u \rVert_{\ell+1} + \lVert p \rVert_{\ell} \right), \quad \ell = 1, 2
$$

The standard Taylor-Hood element of degree $\ell$ in the velocity approximation and $\ell-1$ in the pressure approximation would yield expected approximation error convergence rates of:


$$
\lVert \mathbf{u} - \mathbf{u}_h \rVert_{L^2} \sim O(h^{\ell + 1})
$$
$$
\lVert \mathbf{u} - \mathbf{u}_h \rVert_{H^1} \sim O(h^\ell)
$$
$$
\lVert p - p_h \rVert_{L^2} \sim O(h^\ell)
$$
$$
\lVert p - p_h \rVert_{H^1} \sim O(h^{\ell - 1})
$$
